Inserting the training and testing data into numpy to deal them as array

```python
import numpy as np

# Training and Test Data
train = np.array([
    [1.00, 0.08, 0.72, 1.0],
    [1.00, 0.10, 1.00, 1.0],
    [1.00, 0.26, 0.58, 1.0],
    [1.00, 0.35, 0.95, 0.0],
    [1.00, 0.45, 0.15, 1.0],
    [1.00, 0.60, 0.30, 1.0],
    [1.00, 0.70, 0.65, 0.0],
    [1.00, 0.92, 0.45, 0.0]
])

test = np.array([
    [1.00, 0.42, 0.85, 0.0],
    [1.00, 0.65, 0.55, 0.0],
    [1.00, 0.20, 0.30, 1.0],
    [1.00, 0.20, 1.00, 0.0],
    [1.00, 0.85, 0.10, 1.0]
])
```

Define the learning rate and epochs values

```python

learning_rates = [0.1, 0.2, 0.3]
epochs = 15 # i make it 15 to show how the bias wiil change 
```

first : i will make a function to perform the Perceptron algorithm.

second :i will make for loop on the epochs of train data to loop as my selected epochs .

third : i will updates weights and use the activation function.

forth : i will Update rules.

fifth : i will print weights and bias after each epoch.

sixth : now i will use the test data .

seventh : i will Prediction the test.

eighth : i will Check if prediction is working in the right way .

ninth : i will Calculate the loss for tha test for mean square error.

tenth : i will Calculate accuracy and MSE.

eleventh : i will print the results .

finally : i will call the function . 


```python
def perceptron_algorithm(train, test, learning_rate, epochs):
    weights = np.zeros(3)
    bias = 0
    
    print(f"\nLearning Rate: {learning_rate}")

    # Training 
    for epoch in range(epochs):
        for sample in train:
            inputs = sample[:3]
            target = sample[3]

            # Calculate weighted sum
            weight = np.dot(weights, inputs) + bias
            
            # Activation function 
            output = 1 if weight > 0 else 0
            
            # Update rule
            error = target - output
            weights += learning_rate * error * inputs
            bias += learning_rate * error

        print(f"Epoch {epoch + 1}: Weights = {weights}, Bias = {bias}")

    # Testing 
    predictions = 0
    loss = 0

    for sample in test:
        inputs = sample[:3]
        target = sample[3]

        # Prediction the test
        weighted = np.dot(weights, inputs) + bias
        output = 1 if weighted> 0 else 0

        # Check if prediction is working in the right way .
        if output == target:
            predictions += 1 ## to calculate the correct pridict for accuracy and MSE.

        # Calculate the loss for tha test for mean square error
        loss += (target - output) ** 2

    # Calculate accuracy and MSE
    accuracy = (predictions / len(test)) * 100
    mse = loss / len(test)

    # Print the results
    print(f"\nFinal Learning Rate: {learning_rate}")
    print(f"Weights: {weights}")
    print(f"Bias: {bias}")
    print(f"Accuracy: {accuracy}%")
    print(f"(MSE): {mse}\n")

# Loop over each learning rate and train the model
for learning_rate in learning_rates:
    perceptron_algorithm(train, test, learning_rate, epochs)
```
